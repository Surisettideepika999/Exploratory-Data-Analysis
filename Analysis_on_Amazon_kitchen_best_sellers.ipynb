```python
req_col=['title', 'brand', 'stars', 'reviewsCount',
       'price/currency', 'price/value']
```

```python
df[req_col].head()
```

```python
df_copy=df[req_col]
```

```python
df_copy.nunique()
```

```python
df_copy.drop('price/currency',axis=1,inplace =True)
```

```python
best_seller=df.brand.value_counts().sort_values(ascending=False).head(10)
best_seller
```

```python
sns.barplot(best_seller)
plt.xticks(rotation=75)
plt.plot()
```

we can see that zulay kitchen appeared more in the market

```python
#Identifying top 10 brands with highest reviews

brand_reviews = df_copy.groupby("brand")[["reviewsCount"]].sum().sort_values(ascending=False,by="reviewsCount").head(10)
brand_reviews
```

```python
sns.barplot(y=brand_reviews['reviewsCount'],x=brand_reviews.index)
plt.xticks(rotation=75)
plt.xlabel("Brand")
plt.ylabel("reviews")
plt.title("Top 10 brands with highest reviews")
plt.show()
```

Again Zulay kitchen occupied the first position in the rating also

```python
#Identifying bottom 10 brands with least reviews

brand_reviews_low = df_copy.groupby("brand")[["reviewsCount"]].sum().sort_values(by="reviewsCount").head(10)
brand_reviews_low
```

```python
sns.barplot(y=brand_reviews_low['reviewsCount'],x=brand_reviews_low.index)
plt.xticks(rotation=75)
plt.xlabel("Brand")
plt.ylabel("reviews")
plt.title("Bottom 10 brands with least reviews")
plt.show()
```

Ruby Horsepower is the company with least reviews

```python
top_comp=brand_reviews.index
```

```python
top_comp_data =  df_copy[df_copy["brand"].isin(top_comp)]
```

```python
#highest prices in top companies

top_highest_price = top_comp_data.groupby("brand")[["price/value"]].max()

#least prices in top companies

top_least_price = top_comp_data.groupby("brand")[["price/value"]].min()

#average price in top companies

top_avg_price = top_comp_data.groupby("brand")[["price/value"]].mean()
```

```python
top_highest_price
```

```python
bottom_comp=brand_reviews_low.index
bottom_comp_data =  df_copy[df_copy["brand"].isin(bottom_comp)]
```

```python
#highest prices in bottom companies

bottom_highest_price = bottom_comp_data.groupby("brand")[["price/value"]].max()

#least prices in bottom companies

bottom_least_price = bottom_comp_data.groupby("brand")[["bottom_least_price"]].min()

#average price in bottom companies

bottom_avg_price = bottom_comp_data.groupby("brand")[["price/value"]].mean()
```

```python
col = {
    'top_highest_price': top_highest_price,
    'bottom_highest_price': bottom_highest_price,
    'top_least_price': top_least_price,
    'bottom_least_price': bottom_least_price,
    'top_avg_price': top_avg_price,
    'bottom_avg_price': bottom_avg_price
}

fig,axes=plt.subplots(len(col)//2,2,figsize=(10,10))
axes=axes.flatten()

for i,(name,col) in enumerate(col.items()):
  sns.barplot(x=col['price/value'],y=col.index,ax=axes[i]);
  axes[i].set_title(name)
plt.tight_layout()
plt.show()
```

```python
top_comp_stars=df_copy[df['brand'].isin(top_comp)]
stars_by_brand = top_comp_stars.groupby(['brand', 'stars']).size().reset_index(name='count')

# Find the most frequent star rating for each brand
idx = stars_by_brand.groupby('brand')['count'].idxmax()
most_frequent_stars = stars_by_brand.loc[idx]

print(most_frequent_stars)
```

```python
bottom_comp_stars=df_copy[df['brand'].isin(bottom_comp)]
stars_by_brand = bottom_comp_stars.groupby(['brand', 'stars']).size().reset_index(name='count')

# Find the most frequent star rating for each brand
idx = stars_by_brand.groupby('brand')['count'].idxmax()
most_frequent_stars = stars_by_brand.loc[idx]

print(most_frequent_stars)
```
